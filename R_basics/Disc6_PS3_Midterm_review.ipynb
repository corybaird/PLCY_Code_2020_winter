## 3.3 Z-score

### 3.3.1 Plot normal distribution: $\mu=100, \sigma=5$

```r
#Generate data
normal_hundred_five = rnorm(1000, mean = 100, sd = 5)
#Plot data
hist(
    normal_hundred_five ,#Data
    col='lightblue', #Color
    breaks=10
)
```

### 3.3.2 Standardize the above graph in 3.3.1

$$z = \frac{X-\mu}{\sigma}$$

```r
normal_hundred_five_mean = mean(normal_hundred_five)
normal_hundred_five_sd = sd(normal_hundred_five)
```

```r
z = ((normal_hundred_five-normal_hundred_five_mean)/(normal_hundred_five_sd))
```

```r
hist(
    z,#Data
    col='lightblue', #Color
)
abline(v= sd(z), 
       col="red")
text(0, 100,
     "68% of data lies below the red line",
    col='red')
```

# 5. Sample mean



## 5.1 Calculate sd

$$s^2 = \frac{1}{n-1}\sum^{n}_{i=1}(x_i-\bar{x})^2$$

###  5.1.1 Create random xis

```r
xi = sample.int(10, 10)
xi
```

### 5.1.2 Convert the xis into dataframe

```r
xi_df = as_tibble(xi) %>% rename(xi=value)
xi_df
```

### 5.1.3 Calculate deviations from mean

$$x_i-\bar{x}$$

```r
# You can create a new column using the following code: Dataframe$New_column_name
xi_df$xi_mean = mean(xi_df$xi)
xi_df %>% head(2)
```

```r
xi_df$deviation = (xi_df$xi-xi_df$xi_mean)^2
xi_df %>% head(2)
```

### 5.1.4 Sum the deviations

```r
xi_df$xi_sum = sum(xi_df$deviation)
xi_df %>% head(3)
```

### 5.1.5 Calculate n (this is simply the length of data a.k.a number of observations

```r
xi_df$n = length(xi_df$xi)
xi_df %>% head(3)
```

### 5.1.6 Calculate the variance and standard deviation

```r
xi_df$variance = xi_df$xi_sum/(xi_df$n-1)
xi_df %>% head(3)
```

```r
xi_df$standard_deviation = sqrt(xi_df$variance)
xi_df %>% head(3)
```

## 5.2 Law of large numbers

### 5.2.1 Create coin toss function

```r
head_counter <- function(times = 1) {
    #Coin input
    coin <- c('heads', 'tails')
    prob = c(.5,.5)
    #Sample function
    sample = sample(coin, size = times, replace = TRUE, prob = prob)
    #Count number of heads
    head_count = sum(as.numeric(sample=='heads'))
    #Output head count
    return (head_count)
}
```

```r
head_counter(10)
```

### 5.2.2 Simulate 1000 coin flips

```r
all_experiments = vector(mode = "list", length = 1000)

for (toss in 1:1000){
    all_experiments[toss] = head_counter(toss)/toss
}
```

```r
x = c(1:1000)
y = all_experiments 
plot(x,
     y,
     xlab='Toss number',
     ylab=' Proportion of Heads',
     main='Law of large numbers example',
     col='red'
    )
```

## 5.3 Central limit theorem

```r
n_100 = vector(mode = "list", length = 100)
n_20 = vector(mode = "list", length = 20)
n_5 = vector(mode = "list", length = 5)

for (toss in 1:100){
    n_100[toss] = head_counter(toss)/toss
}

for (toss in 1:20){
    n_20[toss] = head_counter(toss)/toss
}

for (toss in 1:5){
    n_5[toss] = head_counter(toss)/toss
}
```

```r
hist(as.numeric(n_100))
hist(as.numeric(n_20), col='lightblue', add=TRUE)
hist(as.numeric(n_5),col='lightgreen', add=TRUE)
```
